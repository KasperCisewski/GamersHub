Extracting Genre and Date. Splitting them into two parts.

```python
def getGenAndDate():
    for game in game_table('p',{'class' : 'opis-b'}):
        genre,Date = game.text.split(' ', 1)
        gameGenre.append(genre)
        gameDate.append(Date)
```

Extracting short description about game.

```python
def getDesc():
    for game in game_table('p', class_=""):
        gameDesc.append(game.text)
```

Extracting Photo Url from game.

```python
def getURL():
    for game in game_table('img'):
        gamePhotoUrl.append(game['data-src'])
        if not game['data-src']:
            gamePhotoUrl.append('empty')
```

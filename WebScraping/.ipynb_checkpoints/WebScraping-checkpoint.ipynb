Extracting short description about game.

```python
def getDesc():
    flag = 0
    for game in game_table('p'):
        if game.text:
            li = game.text.splitlines()
            if not li:
                         gameDesc.append('empty')
            for element in li:
                if flag == 1:
                    gameDesc.append(element)                    
                if flag == 3:
                    flag = 0
                flag += 1
```

Extracting Photo Url from game.

```python
def getURL():
    for game in game_table('img'):
        gamePhotoUrl.append(game['data-src'])
        if not game['data-src']:
            gamePhotoUrl.append('empty')
```
